```python
import numpy as np # linear algebra, data manipulation
import pandas as pd # data processing, # Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
# Manipulating dates and time
from datetime import datetime
```

```python
#importing files and exploring the data
train_data=pd.read_excel("train_users_2 (2).xlsx")
test_data =pd.read_excel("test_users.xlsx")
```

```python
# compare difference between train and test csv files columns
print(np.setdiff1d(train_data.columns,test_data.columns, assume_unique=True))
```

```python
# more info about training data
print("The shape of training data is: ",train_data.shape)
print("The shape of teast data is: ",test_data.shape)

train_data.info()
''' notice that there are a lot of null values at 'date_first_booking', 'age','first_affiliate_tracked'
and also Columns like :date_account_created ,timestamp_first_active are dates but not in the date format'''
```

```python
test_data.info()
```

```python
# concatenate train and test data together, drop country destination, then check for null values
concat_data = pd.concat([train_data, test_data], axis=0)
concat_data.reset_index(drop=True)
print("The shape of all data is: ",concat_data.shape)
concat_data.info()


```

```python
concat_data.head()
```

```python
# Converting to date time format
concat_data['date_account_created'] = pd.to_datetime(train_data['date_account_created'])
concat_data['date_first_booking'] = pd.to_datetime(train_data['date_first_booking'])
concat_data['timestamp_first_active'] = pd.to_datetime(train_data['timestamp_first_active'], format='%Y%m%d%H%M%S')
```

```python
# investigate null percentage and visualize it
plt.figure(figsize = (20,8))
sns.heatmap(concat_data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
print('Date first booking null value percentage : ',(concat_data['date_first_booking'].isnull().sum()/len(concat_data['date_first_booking'].isnull())*100),'%\n')
print('Age null value percentage : ',(concat_data['age'].isnull().sum()/len(concat_data['age'].isnull())*100),'%\n')
print('first_affiliate_tracked null value percentage : ',(concat_data['first_affiliate_tracked'].isnull().sum()/len(concat_data['first_affiliate_tracked'].isnull())*100),'%\n')
```

```python
concat_data.isnull().sum()
```

```python
#explore numerical columns, starting with age
concat_data.age.describe()
```

```python
plt.figure(figsize=(12,8))
sns.distplot(concat_data.age.dropna())
plt.title('Age Distrubtion')
plt.show()
```

```python
#Excluding nan, less than 15 and grater than 95 values
data_a = concat_data
data_a.loc[(concat_data.age > 105) | (concat_data.age < 15), 'age'] = np.nan
plt.figure(figsize=(12,6))
sns.histplot(data_a.age.dropna())
plt.title('Age Distrubtion excluding nan and outliers (keeping only ages between 15 to 95)')
plt.show()
```

```python
# compare age to countries

data_a.loc[(concat_data.age > 105) | (concat_data.age < 15), 'age'] = np.nan
plt.figure(figsize=(12,6))
sns.boxplot(x='country_destination',y='age', data=data_a.dropna())
plt.xlabel('destination countries')
plt.title('Age vs. destination countries')
plt.show()
```
